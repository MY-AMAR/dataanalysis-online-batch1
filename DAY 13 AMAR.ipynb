### Histogram plot

```python
stud_1 = np.random.randint(18,45,100)
stud_2 = np.random.randint(15,40,100)
plt.hist(stud_1)
plt.xlabel('age category')
plt.ylabel("no of students")
plt.show()
```

```python
plt.hist(stud_1, bins=5)
```

```python
bins=[15,20,25,30,35,40,45]
plt.hist(stud_1,bins)
```

```python
bins=[15,20,25,30,35,40,45]
plt.hist(stud_1,bins, histtype='bar', orientation='vertical', color='m')
plt.show()
```

```python
help(plt.hist)
```

```python
from matplotlib import style
style.use("ggplot")
plt.figure(figsize=(16,9))
plt.hist([stud_1, stud_2], bins, histtype ='step', orientation='vertical', color =['g','y'])
plt.title("histogram example")
plt.xlabel("age category")
plt.ylabel("no of students")
plt.show()
```

### pie chart

```python
x=[15,25,30,5,10]
plt.pie(x)
plt.show()
```

```python
labels=['very likely ', 'likely','unsure','veryunlikely', 'sure']
plt.pie(x,labels=labels)
plt.show()
```

```python
colors=['r','b','g','c','m']
plt.pie(x,labels=labels,colors=colors, autopct="%.0f%%")
plt.title("survey report")
plt.show()
```

```python
colors=['r','b','g','c','m']
explode=[0,0,0,0,0.2]
plt.pie(x,labels=labels,colors=colors, explode=explode, autopct="%.0f%%")
plt.title("survey report")
plt.show()
```

```python
plt.pie(x,labels=labels,colors=colors, explode=explode, autopct="%.0f%%", shadow=True)
plt.title("survey report")
plt.show()
```

```python
plt.pie(x,labels=labels,colors=colors, explode=explode, autopct="%.0f%%", shadow=True, startangle=180)
plt.title("survey report")
plt.show()
```

*** stack plot ***

```python
x=[1,2,3,4,5]
y1=[6,7,8,9,10]
y2=[10,20,30,40,50]
y3=[15,25,39,49,51]
plt.stackplot(x,y1,y2,y3)
plt.show()
```

```python
x=[1,2,3,4,5]
y_values={
"y1":[6,7,8,9,10],
"y2":[10,20,30,40,50],
"y3":[15,25,39,49,51]}
plt.stackplot(x,y_values.values())
plt.show()
```

```python
plt.stackplot(x,y_values.values(),labels=y_values.keys())
plt.legend(loc="upper left")
```
