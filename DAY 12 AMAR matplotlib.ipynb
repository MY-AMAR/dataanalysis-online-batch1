```python
days=[1,2,3,4,5]
temp=[36.6,37,37.7,38,40.1]
vij_temp=[39.9,40,40.4,46.7,49.8]
plt.plot(days,temp, 'm^--', linewidth=3, label='guntur temp')
plt.plot(days,vij_temp, 'yo:', linewidth=3, markersize=10, label='vij temp')
plt.title("Temperatures of vij and guntur")
plt.xlabel("days", fontsize=15)
plt.ylabel("temperature",fontsize=20)
plt.legend(loc=4)
plt.show()
```

```python

classes=["python","R", "AI", 'ML', 'Datascience']
c1_stud=[30,10,20,25,10]
c2_stud=[40,20,5,40,50]
C3_stud=[35,49,10,29,70]

cls_index=np.arange(len(classes))
width=0.2
plt.bar(cls_index,c1_stud,width,color='b')
plt.bar(cls_index+width,c2_stud,width,color='g')
plt.bar(cls_index+width+width,C3_stud,width,color='y')
```

```python
cls_index=np.arange(len(classes))
width=0.2
plt.bar(cls_index,c1_stud,width,color='b')
plt.bar(cls_index+width,c2_stud,width,color='g')
plt.bar(cls_index+width+width,C3_stud,width,color='y')
plt.xticks(cls_index+width,classes)
```

```python
cls_index=np.arange(len(classes))
width=0.2
plt.bar(cls_index,c1_stud,width,color='b')
plt.bar(cls_index+width,c2_stud,width,color='g')
plt.bar(cls_index+width+width,C3_stud,width,color='y')
plt.xticks(cls_index+width,classes,rotation=20)
```
