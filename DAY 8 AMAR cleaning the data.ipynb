## Handling missing value using Pandas
* mainly 4 methods are there for identifying and manipulate missing values 
 1. isnull() Presense of missing values, returns boolean data
 2. notnull() opposite to isnull()
 3. dropna()  It removes null values
 4. filina() Fills the missing values by specified values

### Replace null values
* If the data is numerical data set Replace the Null values with 0 or mean() or median() or any other values 
* If the data is a categorical data, replace the null values with the most occured name or any user specified name
* fillna()

```python
df.head(2)
```

### ffill and bfill

```python
dict2= {1:[111,112,np.NaN,114,115,np.NaN],
       2: [222,np.NaN,223,224,225,226],
       3:[np.NaN,332,np.NaN, 334,np.NaN,336]}
df2= pd.DataFrame(dict2)
```

```python
df2
```

```python
#  ffill : front fill
# apply ffill for the first column
```

```python
df2[1].fillna(method='ffill',inplace= True) # NaN is replaced by before value
df2 
```

```python
help(df2.fillna)
```

```python
df2[2].fillna(method='bfill',inplace= True) # NaN is replaced by before value
df2 
```

```python
df2.fillna(0,inplace =  True)
df2
```
