# Creating DataFrame

```python
import pandas as pd
l=[11,22,33,44,55,66]
df=pd.DataFrame(l)
df
```

```python
# Converting a list into Dataframe
l=[[1,2,3],[4,5,6],[7,8,9]]
df1=pd.DataFrame(l)
df1
```

```python
df1.index=["a","b","c"] # reindexing rows
df1
```

```python
df1.columns=["A","B","C"] # reindexing columns
df1
```

```python
# creating dataframe using pandas series
s= pd.Series([77,88,33,22,44], index=[1,2,3,4,5])
df=pd.DataFrame(s)
df
```

```python
# creating dataframe using Dict
d={"a":90,"b":100,"c":300}
df=pd.DataFrame(d,index=[1,2,3])
df
```

```python
d={"a":pd.Series([10,20,30,40],index=[1,2,3,4])}
df=pd.DataFrame(d)
df
```

```python
d={"a":pd.Series([10,20,30,40],index=[1,2,3,4]),
   "b":pd.Series([11,12,13],index=[1,2,3]),
   "c":pd.Series([25,25,35],index=[1,2,3])
  }
df=pd.DataFrame(d)
df
```

```python
d={"a":pd.Series([10,20,30,40],index=[1,2,3,4]),
   "b":pd.Series([11,12,13],index=[1,2,3]),
   "c":pd.Series([25,25,35],index=[2,3,5])
  }
df=pd.DataFrame(d)
df
```

# Rows having index values and columns having meaningful name/ information about data

```python
f=[[i**2 for i in range(10,21)],[j**3 for j in range(10,21)]]
pd.DataFrame(f) #, index=['squares','cubes'],columns=range(10,21))
```

```python
df = pd.DataFrame([[i**2 for i in range(10, 21)], [i**3 for i in range(10, 21)]], index = ["squares", "cubes"])
df.columns = range(10, 21)
df
```

```python
data=[{"squares":n**2,"cubes":n**3} for n in range(10,21)]
pd.DataFrame(data)
```

```python
data=[{"squares":n**2,"cubes":n**3} for n in range(10,21)]
pd.DataFrame(data, index=range(10,21))
```

```python
d={"a":pd.Series([i**2 for i in range(1,5)],index=[1,2,3,4]),
   "b":pd.Series([i**3 for i in range(1,5)],index=[1,2,3,4]),
  }
df=pd.DataFrame(d)
df
```

```python
d1=pd.DataFrame({"emp":pd.Series(["a","b","c","d"],index=[1,2,3,4]), 
                 "yr":pd.Series(["2000","2001","2002","2003"],index=[1,2,3,4])
                 })
print(d1)
```

```python
d2=pd.DataFrame({"emp":pd.Series(["a","b","c","e"],index=[1,2,3,4]), 
                 "dept":pd.Series(["hr","op","ma","s"],index=[1,2,3,4])
                 })
print(d2)
```

```python
d1.merge(d2)
```

```python
d1.merge(d2,how="left")
```

```python
d1.merge(d2,how="right")
```

```python
df1 = pd.DataFrame({'lkey': ['foo', 'bar', 'baz', 'foo'],
                    'value': [1, 2, 3, 5]})
df2 = pd.DataFrame({'rkey': ['foo', 'bar', 'baz', 'foo'],
                  'value': [5, 6, 7, 8]})
print(df1)
print(df2)
```

```python
df1.merge(df2)
```

```python
df1.merge(df2,how="left")
```

```python
df1.merge(df2,how="right")
```
